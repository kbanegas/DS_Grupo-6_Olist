```python
# Implemento funcin que permite detectar Outliers, la cual retorna una lista, donde el primer elemento es el porcentaje que respresentan los
# Outliers encontrados sobre el total de datos de la columna en cuestin, y el segun valor de la lista es la ubicacin de los Outliers.
def calculo_porcentaje_outliers(df, columna):
    q1 = df[columna].quantile(0.25)
    q3 = df[columna].quantile(0.75)
    rango = q3 - q1
    bi = (q1 - 1.5 * rango)
    bs = (q3 + 1.5 * rango)
    ubic_out = ((df[columna] < bi) | (df[columna] > bs))
    outliers = df[ubic_out].shape[0]
    total_reg = df.shape[0]
    porcentaje = outliers / total_reg
    resultado = [round(porcentaje * 100, 3), ubic_out]
    return resultado
```

```python
# Identificando outliers en las columnas:
lista_resultado = calculo_porcentaje_outliers(df_order_items, "price")
print("El porcentaje de otliers en los pagos es de:", lista_resultado[0], "%")
```

```python
lista_resultado = calculo_porcentaje_outliers(df_order_items, "freight_value")
print("El porcentaje de otliers en los pagos es de:", lista_resultado[0], "%")
```

```python
# Lo outliers son los siguientes:
df_order_payments[lista_resultado[1]].sort_values("payment_value")
# Se pueden ver visualizados por orden de menor a mayor segun su valor:
```
