```python
import pandas as pd
import numpy as np
```

```python
#Normalizacion de tabla review 
reviews= pd.read_csv('C:\Henry\Labs\Grupal\DS-Proyecto_Grupal_Olist-main\data\e-comerce_Olist_dataset\olist_order_reviews_dataset.csv') #levanto el dataset
review= reviews[['review_id', 'order_id', 'review_score', 'review_comment_message', 'review_creation_date', 'review_answer_timestamp']] #me quedo con las tablas importantes 
review.rename(columns={'review_id': 'idcomentario', 
                           'order_id': 'idorder', 'review_score':'puntacion', 'review_comment_message':'comentario', 'review_creation_date': 'fecha_comentario', 'review_answer_timestamp':'fecha_respuesta'}, inplace=True)#cambio de nombres de cada tabla
						   
review['comentario'] = review['comentario'].replace(np.nan,'Sin Dato') # Reemplazo nulos en comentario por sin dato

#reviewcsv = review.to_csv('review.csv', index = True) #exporto el csv
```
